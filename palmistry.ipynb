```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Load dataset
# Assuming 'images' is a directory containing the palm and hand images
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_generator = datagen.flow_from_directory(
    '/kaggle/input/hands-and-palm-images-dataset',
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

validation_generator = datagen.flow_from_directory(
    '/kaggle/input/hands-and-palm-images-dataset',
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

# Build CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(4, activation='softmax')  # 4 units for 4 classes
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(train_generator, validation_data=validation_generator, epochs=1)
```

```python
from PIL import Image
import matplotlib.pyplot as plt
from keras.preprocessing import image
import numpy as np

# Load the image
img_path = "/kaggle/input/hands-and-palm-images-dataset/Hands/Hands/Hand_0000032.jpg"
img = Image.open(img_path)

# Display the image
plt.imshow(img)
plt.axis('off')
plt.show()

# Resize the image
resized_img = img.resize((150, 150))

# Convert resized image to array
img_array = image.img_to_array(resized_img)

# Normalize the image data
img_array = img_array / 255.0

# Reshape the array to match the expected input shape of the model
img_array_reshaped = img_array.reshape(1, 150, 150, 3)

# Make predictions
predictions = model.predict(img_array_reshaped)
```
